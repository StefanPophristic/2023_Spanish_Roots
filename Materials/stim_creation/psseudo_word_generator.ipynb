```python
'''
relative_entropy_calculation.py
Stefan Pophristic
February 15, 2024

Input: {ar_tokens_count.csv, er_tokens_count.csv, ir_tokens_count.csv,
        ar_suffix_count.csv, er_suffix_count.csv, ir_suffix_count.csv}
    6 csv files output of frequencies_4.py *r_tokens_count.csv are csv files with the
    count of all tokens of words that fall into a particular verb conjugation class.
    *r_suffix_count.csv are csv files with the count of suffix endings for a
    particular verb conjugation class across all lemmas/words.

    Note that the input *r_tokens_count.csv files have separate counts based on capitalization 
    of word forms. Ideally this should be changed in frequencies_2.py. Due to time 
    I will include cleaning in this script. 

Output: {relative_entropy.csv}
    csv file with a relative entropy value for each verbal lemma in the corpus.

Relative Entropy calculation from Milin, Filipović-Ðurdević, Moscoso del Pradi martín 2009

D(p||q) = SUM[ p(i) log2[ p(i) / q(i) ] ]
        = sum[ (f(wi) / f(w)) * log2[ (f(wi) / f(w)) (f(ei) / f(e)) ] ]

p = word form (root + ending)
q = Inflectional Class Distribution
f(wi) = frequency (raw count) of a word form (root + suffix)
f(w) = frequency (raw count) of all word forms of a lemma
f(ei) = summed frequency (raw count) of all words with a particular suffix in a given IC
f(e) = summed frequency (raw count) of all words in the IC
'''
```

```python
########
#### Complete checks of dataframes and gather metadata 
########


# Function that checks if dataframe has capitalized values 
def has_capitalized(df):
    for column in ["word", 'lemma']:
        # Check each value in the column for capitalization
        if any(value.isupper() for value in df[column]):
            return True
    return False

# Check if any of the imported dataframes have capitalized values
for x in [ar_words, er_words, ir_words]:
    if has_capitalized(x):
        print("DataFrame has capitalized values.")
    else:
        print("DataFrame does not have capitalized values.")

# Check if suffix and word dataframes have same total number of tokens
def get_total_token_count(df):
    return(df['count'].sum())

print("ar token number check: " + str(get_total_token_count(ar_words) == get_total_token_count(ar_suffixes)))
print("er token number check: " + str(get_total_token_count(er_words) == get_total_token_count(er_suffixes)))
print("ir token number check: " + str(get_total_token_count(ir_words) == get_total_token_count(ir_suffixes)))

# get total suffix count not includeing the "other" suffixes 
# i.e. suffixes that are mistakes, or that we don't care about 
ar_suffixes_total_count = ar_suffixes.loc[ar_suffixes['suffix'] != 'other', 'count'].sum()
print(ar_suffixes_total_count)

er_suffixes_total_count = er_suffixes.loc[er_suffixes['suffix'] != 'other', 'count'].sum()
print(er_suffixes_total_count)

ir_suffixes_total_count = ir_suffixes.loc[ir_suffixes['suffix'] != 'other', 'count'].sum()
print(ir_suffixes_total_count)
```

```python
########
#### Define Functions
########


def get_IC(IC_input):
    if IC_input == "ar" :
        return(ar_words)
    elif IC_input == "er" :
        return(er_words)
    else:
        return(ir_words)

def get_word_count(word_input, IC_input):
    try:
        tokens = get_IC(IC_input)
        count_value = tokens[tokens['word'] == word_input]["count"].iloc[0]
        return count_value
    except IndexError:
        # Handle the case where the value is not found
        return pd.NA

def get_lemma_count(lemma_input, IC_input):
    try:
        tokens = get_IC(IC_input)
        count_value = tokens[tokens['lemma'] == lemma_input]["count"].sum()
        return count_value
    except IndexError:
        # Handle the case where the value is not found
        return pd.NA

# Return log 10 count of a word form
# def get_word_p(word_input, IC):
#     prob = math.log10(get_word_count(word_input, IC))
#     return prob
def get_word_p(word_input, IC):
    try:
        word_count = get_word_count(word_input, IC)
        if word_count is None or math.isnan(word_count) or math.isinf(word_count):
            return 0
        
        prob = math.log10(word_count)
        return prob
    except (ValueError, TypeError):
        # Return 0 for any error without printing the error message
        return 0

# Return log 10 count of a lemma
def get_lemma_p(lemma_input, IC):
    prob = math.log10(get_lemma_count(lemma_input, IC))
    return prob

# Return the transition probability between a word and its whole suffix complex
# Input: 
#   lemma_input: (str) lemma of the word
#   word_input: (str) word form with desired suffix
# Output: 
#   Transition probability between stem and suffix
# def get_TP(lemma_input, word_input):
#     final_char = lemma_input[-2:]
#     if final_char in ["ar", "er", "ir"]:
#         return(get_word_p(word_input, final_char) / get_lemma_p(lemma_input, final_char))
#     else:
#         raise ValueError(final_char + " is not a valid infinitival suffix.")
def get_TP(lemma_input, word_input):
    try:
        final_char = lemma_input[-2:]
        if final_char in ["ar", "er", "ir"]:
            word_p = get_word_p(word_input, final_char)
            lemma_p = get_lemma_p(lemma_input, final_char)
            return word_p / lemma_p
        else:
            raise ValueError(final_char + " is not a valid infinitival suffix.")
    except (ValueError, TypeError) as e:
        # Handle errors raised by the embedded functions
        return 0
# NAType

# Return the suffix from spanish_suffixes variable that best matches word form
# Input: 
#   word: (str) word form with desired suffix
# Output: 
#   (str) suffix as defined in spanish_suffixes
def best_matching_suffix(word):
    best_suffix = None
    max_suffix_length = 0

    for suffix in spanish_suffixes:
        if word.endswith(suffix) and len(suffix) > max_suffix_length:
            best_suffix = suffix
            max_suffix_length = len(suffix)

    return best_suffix


# Return the suffix from spanish_suffixes variable that best matches word form
# Input: 
#   lemma: (str) lemma form of word (must end in /ar/, /er/, or /ir/)
# Output: 
#   suffix_list: dataframe with all suffixes of that class and their token counts
#   word_list: dataframe with all word tokens, lemmas, and their token counts
#   total_suffix_count: integer of total tokens within a class

def get_IC_RE(input_lemma):
    
    final_char = input_lemma[-2:]

    if final_char == "ar":
        suffix_list = ar_suffixes
        word_list = ar_words
        total_suffix_count = ar_suffixes_total_count
    elif final_char == "er":
        suffix_list = er_suffixes
        word_list = er_words
        total_suffix_count = er_suffixes_total_count
    elif final_char == "ir" :
        suffix_list = ir_suffixes
        word_list = ir_words
        total_suffix_count = ir_suffixes_total_count
    else:
        raise ValueError(final_char + " is not a valid infinitival suffix.")
        
    return(suffix_list, word_list, total_suffix_count)

# Return relative entropy value of a lemma
# Input:
#   input_lemma: (str) lemma form of word (must end in /ar/, /er/, or /ir/)
# Output: 
#   (float) Relative entropy value as calculated according to Milin et al. 2009
def get_RE(input_lemma): 
    # Relative Entropy value is a summation
    RE = 0
    # test_sum = 0
    # test_IC_sum = 0
    
    suffix_list, word_list, total_suffix_count = get_IC_RE(input_lemma)
    
    # get count of all forms of that lemma
    filtered_lemma = word_list[word_list["lemma"] == input_lemma]

    # Total sum variable
    lemma_count = filtered_lemma['count'].sum()

    # loop through all forms
    for w in filtered_lemma["word"]:
        # print(w)
        # Get specific form count
        form = filtered_lemma[filtered_lemma["word"] == w]
        form_count = int(form["count"].iloc[0])

        # Match word form with form in the suffix table
        word_suffix = best_matching_suffix(w)
        # print(word_suffix)

        # calculate relative entropy value for this specific word form + IC suffix
        word_freq = form_count / lemma_count
        # test_sum += word_freq
        
        if word_suffix == None: 
            RE_word = 0
        else:
            suffix_count = suffix_list.loc[suffix_list['suffix'] == word_suffix, 'count'].iloc[0]
            suffix_freq = int(suffix_count) / total_suffix_count
            # test_IC_sum += suffix_freq
    
            if suffix_freq == 0:
                RE_word = 0  # Assign a default value or handle it as needed
            else:
                RE_word = word_freq * math.log2(word_freq / suffix_freq)
        # add value to sum variable for final relative entropy variable
        RE += RE_word
    return(RE)

# Return dataframe with all forms of a lemma 
def get_lemma_forms(input_lemma):
    suffix_list, word_list, total_suffix_count = get_IC_RE(input_lemma)

    words = []
    suffixes = []
    
    # get count of all forms of that lemma
    filtered_lemma = word_list[word_list["lemma"] == input_lemma]

    # loop through all forms
    for w in filtered_lemma["word"]:
        # Match word form with form in the suffix table
        words.append(w)
        word_suffix = best_matching_suffix(w)
        suffixes.append(word_suffix)
    
    output_df = pd.DataFrame({'word': words, 'suffix': suffixes})
    output_df = output_df.sort_values(by='suffix')
    return output_df

# Replace all tokens of one form with another in the inflectional class
# Used to correct for spelling errors
# Input:
#   form_bad: (str) string of the word form that you want to replace
#   form_good: (str) string of the word form that should replace form_bad
#   IC: (str) {"ar", "er", "ir"} specifying word form's class 
def fix_form(form_bad, form_good, IC):
    global ar_words, er_words, ir_words

    word_df = get_IC(IC)

    bad_row_index = word_df.index[word_df['word'] == form_bad].tolist()
    good_row_index = word_df.index[word_df['word'] == form_good].tolist()

    if bad_row_index and good_row_index:
        bad_row_index = bad_row_index[0]
        good_row_index = good_row_index[0]

        word_df.at[good_row_index, 'count'] = word_df.at[good_row_index, 'count'] + word_df.at[bad_row_index, 'count']
        word_df = word_df.drop(bad_row_index)
    
    if IC == "ar":
        ar_words = word_df
    elif IC == "er":
        er_words = word_df
    elif IC == "ir":
        ir_words = word_df

# def fix_suffix(word_form, wrong_suffix_, good_suffix, IC):
#     # take count of the word form 
#     # add it to the good_suffix
#     # take it off of the bad_suffix

    # Does this affect anything else??? --> i.e. you pull the suffixes in the calculations,
    # you might need to hard code these exceptions into the function that returns suffixes. 
    # basically like, oh if the word form you got is "case" actually just return "e" not "se"

# Delete specified word form 
# Used to correct for wrong parses in the corpus
# Input:
#   form_bad: (str) string of the word form that you want to delete
#   ipnut_lemma: (str) lemma form associated with the word form you wish to delete
#   wrong_suffix: (str) the parse of the suffix from the word form you wish to delete
#   IC: (str) {"ar", "er", "ir"} specifying word form's class 

def drop_form(form_bad, input_lemma, wrong_suffix, IC):
    global ar_words, er_words, ir_words, ar_suffixes, er_suffixes, ir_suffixes

    suffix_list, word_list, total_suffix_count = get_IC_RE(input_lemma)

    # Find the index of the row of the word you want to drop
    bad_row_index = word_list.index[word_list['word'] == form_bad].tolist()
    
    if not bad_row_index:
        raise ValueError(f"Word '{form_bad}' not found in the DataFrame.")
        
    bad_row_index = bad_row_index[0]

    # get total number of tokens of that word
    bad_row_count = word_list.at[bad_row_index, 'count']

    # drop the bad row
    word_list = word_list.drop(bad_row_index)

    # subtract the tokens of the dropped word from the suffixes dataframe
    new_value = suffix_list[suffix_list["suffix"] == wrong_suffix]['count'].iloc[0] - bad_row_count
    suffix_list.loc[suffix_list['suffix'] == wrong_suffix, 'count'] = new_value

    # update all of this for the global variables
    if IC == "ar":
        ar_words = word_list
        ar_suffixes = suffix_list
    elif IC == "er":
        er_words = word_list
        er_suffixes = suffix_list
    elif IC == "ir":
        ir_words = word_list
        ir_suffixes = suffix_list



# def get_lemma_words(input_lemma):
#     suffix_list, word_list, total_suffix_count = get_IC_RE(input_lemma)

#     output = []

#     # get count of all forms of that lemma
#     filtered_lemma = word_list[word_list["lemma"] == input_lemma]

#     # loop through all forms
#     for w in filtered_lemma["word"]:
#         output.append(w)
#     return(output)


# def print_lemma_suffixes(input_lemma):
#     suffix_list, word_list, total_suffix_count = get_IC_RE(input_lemma)

#     # get count of all forms of that lemma
#     filtered_lemma = word_list[word_list["lemma"] == input_lemma]

#     # loop through all forms
#     for w in filtered_lemma["word"]:
#         # Match word form with form in the suffix table
#         word_suffix = best_matching_suffix(w)
#         output.append(word_suffix)
#     return(output)
```

```python
########
#### Calculate Affix to Affix TP
########
# These are constant regardless of root
# These are calculated over all paradigms


all_suffixes = pd.concat([ar_suffixes, er_suffixes, ir_suffixes]).groupby('suffix', as_index=False).sum()

def sum_suffixes(suffixes_input):
    output = 0
    for x in suffixes_input:
        output += all_suffixes[all_suffixes["suffix"] == x]['count'].iloc[0]
    return output

###
# aba + {s, mos, n}
###
aba_forms_sum = sum_suffixes(["aba", "abas", "ábamos", "abais", "aban"])

abas_TP = math.log10(all_suffixes[all_suffixes["suffix"] == "abas"]['count'].iloc[0]) / math.log10(aba_forms_sum)
abamos_TP = math.log10(all_suffixes[all_suffixes["suffix"] == "ábamos"]['count'].iloc[0]) / math.log10(aba_forms_sum)
aban_TP = math.log10(all_suffixes[all_suffixes["suffix"] == "aban"]['count'].iloc[0]) / math.log10(aba_forms_sum)

print("aba --> s    transition probability: " + str(abas_TP))
print("ába --> mos  transition probability: " + str(abamos_TP))
print("aba --> n    transition probability: " + str(aban_TP))

###
# ía + {s, mos, n}
###
ia_forms_sum = sum_suffixes(["ía", "ías", "íamos", "íais", "ían", 
                             "aría", "arías", "aríamos", "aríais", "arían", 
                             "ería", "erías", "eríamos", "eríais", "erían",
                             "iría", "irías", "iríamos", "iríais", "irían"])

ias_forms_sum = sum_suffixes([ "ías", "arías", "erías", "irías"])
iamos_forms_sum = sum_suffixes([ "íamos", "aríamos", "eríamos", "iríamos"])
ian_forms_sum = sum_suffixes([ "ían", "arían", "erían", "irían"])

ias_TP = math.log10(ias_forms_sum) / math.log10(ia_forms_sum)
iamos_TP = math.log10(iamos_forms_sum) / math.log10(ia_forms_sum)
ian_TP = math.log10(ian_forms_sum) / math.log10(ia_forms_sum)

print("\n")
print("ía --> s     transition probability: " + str(ias_TP))
print("ía --> mos   transition probability: " + str(iamos_TP))
print("ía --> n     transition probability: " + str(ian_TP))

###
# ar + {ías, íamos, ían}
###
ar_forms_sum = sum_suffixes(["aría", "arías", "aríamos", "aríais", "arían", 
                             "aré", "arás", "ará", "aremos", "aréis", "arán",
                             "ara", "aras", "aramos", "áramos", "arais", "aran",
                             "are", "ares", "áremos", "areis", "aren", "ar"])

arias_TP = math.log10(all_suffixes[all_suffixes["suffix"] == "arías"]['count'].iloc[0]) / math.log10(ar_forms_sum)
ariamos_TP = math.log10(all_suffixes[all_suffixes["suffix"] == "aríamos"]['count'].iloc[0]) / math.log10(ar_forms_sum)
arian_TP = math.log10(all_suffixes[all_suffixes["suffix"] == "arían"]['count'].iloc[0]) / math.log10(ar_forms_sum)

print("\n")
print("ar --> ías    transition probability: " + str(arias_TP))
print("ar --> íamos  transition probability: " + str(ariamos_TP))
print("ar --> ían    transition probability: " + str(arian_TP))

###
# er + {ías, íamos, ían}
###
er_forms_sum = sum_suffixes(["ería", "erías", "eríamos", "eríais", "erían", 
                             "eré", "erás", "erá", "eremos", "eréis", "erán",
                             "er"])
# subj. imperfect: "iera", "ieras", "ieramos", "iéramos", "ierais", "ieran"
# Subj. future: "iere", "ieres", "iéremos", "iereis", "ieren"

erias_TP = math.log10(all_suffixes[all_suffixes["suffix"] == "erías"]['count'].iloc[0]) / math.log10(er_forms_sum)
eriamos_TP = math.log10(all_suffixes[all_suffixes["suffix"] == "eríamos"]['count'].iloc[0]) / math.log10(er_forms_sum)
erian_TP = math.log10(all_suffixes[all_suffixes["suffix"] == "erían"]['count'].iloc[0]) / math.log10(er_forms_sum)

print("\n")
print("er --> ías    transition probability: " + str(erias_TP))
print("er --> íamos  transition probability: " + str(eriamos_TP))
print("er --> ían    transition probability: " + str(erian_TP))

###
# ir + {ías, íamos, ían}
###
ir_forms_sum = sum_suffixes(["iría", "irías", "iríamos", "iríais", "irían", 
                             "iré", "irás", "irá", "iremos", "iréis", "irán",
                             "ir"])
# subj. imperfect: "iera", "ieras", "ieramos", "iéramos", "ierais", "ieran"
# Subj. future: "iere", "ieres", "iéremos", "iereis", "ieren"

irias_TP = math.log10(all_suffixes[all_suffixes["suffix"] == "irías"]['count'].iloc[0]) / math.log10(ir_forms_sum)
iriamos_TP = math.log10(all_suffixes[all_suffixes["suffix"] == "iríamos"]['count'].iloc[0]) / math.log10(ir_forms_sum)
irian_TP = math.log10(all_suffixes[all_suffixes["suffix"] == "irían"]['count'].iloc[0]) / math.log10(ir_forms_sum)

print("\n")
print("ir --> ías    transition probability: " + str(irias_TP))
print("ir --> íamos  transition probability: " + str(iriamos_TP))
print("ir --> ían    transition probability: " + str(irian_TP))


###
# Save Values as a Dictionary
###


a1_a2_TP_dict = {'suffix': ['abas', 'ábamos', 'aban', 
                            "ías", "íamos", "ían", 
                            "arías", "aríamos", "arían", 
                            "erías", "eríamos", "erían", 
                            "irías", "iríamos", "irían"], 
                    'TP': [abas_TP, abamos_TP, aban_TP, 
                           ias_TP, iamos_TP, ian_TP,
                           arias_TP, ariamos_TP, arian_TP, 
                           erias_TP, eriamos_TP, erian_TP,
                           irias_TP, iriamos_TP, irian_TP]}

a1_a2_TP_df = pd.DataFrame(a1_a2_TP_dict)

plt.hist(affix_TP_df['TP'], bins=12, edgecolor='black')
plt.xlabel('Transition Probability')
plt.ylabel('Number of Items')
plt.title('Affix to Affix TP distribution')
plt.show()

####
## Subjunctive Calculations
####

###
# Subjunctive: ar + {as, amos, an}
###

# aras_TP = math.log10(all_suffixes[all_suffixes["suffix"] == "aras"]['count'].iloc[0]) / math.log10(ar_forms_sum)
# aramos_TP = math.log10(all_suffixes[all_suffixes["suffix"] == "aramos"]['count'].iloc[0]) / math.log10(ar_forms_sum)
# aran_TP = math.log10(all_suffixes[all_suffixes["suffix"] == "aran"]['count'].iloc[0]) / math.log10(ar_forms_sum)

# print("\n")
# print("ar --> as    transition probability: " + str(aras_TP))
# print("ar --> amos  transition probability: " + str(aramos_TP))
# print("ar --> an    transition probability: " + str(aran_TP))


###
# Present Indicative/Subjunctive: e + {s, mos, n}
###

# subj_es_TP = math.log10(all_suffixes[all_suffixes["suffix"] == "es"]['count'].iloc[0]) / math.log10(ar_forms_sum)
# subj_emos_TP = math.log10(all_suffixes[all_suffixes["suffix"] == "emos"]['count'].iloc[0]) / math.log10(ar_forms_sum)
# subj_en_TP = math.log10(all_suffixes[all_suffixes["suffix"] == "en"]['count'].iloc[0]) / math.log10(ar_forms_sum)

# print("\n")
# print("Subjunctive e --> s    transition probability: " + str(subj_es_TP))
# print("Subjunctive e --> mos  transition probability: " + str(subj_emos_TP))
# print("Subjunctive e --> n   transition probability: " + str(subj_en_TP))
```

```python
########
#### Define Functions that calculate TP of {stem to affix1, stem to affix 2}
########

# Stem to Affix 1 transition probability in the imperfect (stem --> aba; stem --> ía)
def get_stem_a1_imperfect_tp(input_lemma):
    suffix_list, word_list, total_suffix_count = get_IC_RE(input_lemma)
    sum = 0

    lemma_list = word_list[word_list["lemma"] == input_lemma]

    for index, row in lemma_list.iterrows():
        word = row['word']
        # for ar words, count all tokens that end in any of the 'aba' forms
        if input_lemma[-2:] == "ar" and (word.endswith("aba") or word.endswith("abas") or word.endswith("ábamos") or word.endswith("abais") or word.endswith("aban")):
            sum += row['count']
        # for er and ir words, count all tokens that end in one of the ía forms to the exclusion of the conditional present tense forms 
        elif (input_lemma[-2:] == "er" or input_lemma[-2:] == "ir") and (word.endswith("ía") or word.endswith("ías") or word.endswith("íamos") or word.endswith("íais") or word.endswith("ían")) and not (word.endswith("ería") or word.endswith("erías") or word.endswith("eríamos") or word.endswith("eríais") or word.endswith("erían")
                or word.endswith("iría") or word.endswith("irías") or word.endswith("iríamos") or word.endswith("iríais") or word.endswith("irían")):
            sum += row['count']

    output = math.log10(sum)/math.log10(lemma_list['count'].sum())
    return output

# Stem to Affix 1 transition probability in the conditional (stem --> ar, stem --> er, stem --> ir)

# Stem to Affix 1 transition probability for the conditional
#    (stem --> ar, stem --> er, stem --> ir)
#    This function finds all word forms that correspond to input_lemma 
#      and counts all tokens of the words whose forms have the {ar, er, ir} suffix
#      right next to the stem
#    Note: We are exploiting the fact that {ar, er, ir} occur after roots and 
#      nowhere else except for 'ier' forms
#    BEWARE: the root cannot end in /i/ otherwise the form will not be counted
#            the root contain the substrings {ar, er, ir} otherwise those forms will be over counted 
def get_stem_a1_conditional_tp(input_lemma):
    suffix_list, word_list, total_suffix_count = get_IC_RE(input_lemma)
    sum = 0

    lemma_list = word_list[word_list["lemma"] == input_lemma]

    for index, row in lemma_list.iterrows():
        word = row['word']

        # add up tokens of all words that contain 'ar', 'er', and 'ir'
        if (input_lemma[-2:] in word) and ("ier" not in word):
            sum += row['count']

    output = math.log10(sum)/math.log10(lemma_list['count'].sum())
    return output

# Stem to Affix 2 transition probability
#    This function finds all word forms that correspond to input_lemma 
#      and counts all tokens of the words whose forms end in the affix variable
#    Note: there may be more complex affixes that end in the same way as the affix
#      desired. This function only works for the past imperfect and conditional 
#      present tenses
def get_stem_a2_tp(input_lemma, affix):
    try:
        suffix_list, word_list, total_suffix_count = get_IC_RE(input_lemma)
    
        lemma_list = word_list[word_list["lemma"] == input_lemma]
    
        sum = 0
    
        # get sum of a2 forms 
        for index, row in lemma_list.iterrows():
            word = row['word']
            if affix == "s" and word.endswith("s") and not word.endswith("mos") and not word.endswith("is"):
                sum += row['count']
            elif affix != "s" and word.endswith(affix):
                sum += row['count']
    
        output = math.log10(sum)/math.log10(lemma_list[lemma_list["lemma"] == input_lemma]['count'].sum())
        return output
        
    except Exception as e:
        # print(f"An error occurred: {e}")
        return 0
```

```python
########
#### Calculate stem to affix1 and stem to affix 2 probabilities
########

root_tp_dict = {'lemma': [], 'imperfect_stem_a1': [], 
                              'imperfect_stem_a2_s': [],
                              'imperfect_stem_a2_mos': [],
                              'imperfect_stem_a2_n': [],
                              'conditional_stem_a1': [],
                              'conditional_stem_a2_ias': [],
                              'conditional_stem_a2_iamos': [],
                              'conditional_stem_a2_ian': []}

all_stim_lemmas = all_stims_dict['lemma']

for w in all_stim_lemmas:
    input_class = w[-2:]
    
    root_tp_dict['lemma'].append(w)
    root_tp_dict['imperfect_stem_a1'].append(get_stem_a1_imperfect_tp(w))
    root_tp_dict['imperfect_stem_a2_s'].append(get_stem_a2_tp(w, "s"))
    root_tp_dict['imperfect_stem_a2_mos'].append(get_stem_a2_tp(w, "mos"))
    root_tp_dict['imperfect_stem_a2_n'].append(get_stem_a2_tp(w, "n"))
    root_tp_dict['conditional_stem_a1'].append(get_stem_a1_conditional_tp(w))
    root_tp_dict['conditional_stem_a2_ias'].append(get_stem_a2_tp(w, "ías"))
    root_tp_dict['conditional_stem_a2_iamos'].append(get_stem_a2_tp(w, "íamos"))
    root_tp_dict['conditional_stem_a2_ian'].append(get_stem_a2_tp(w, "ían"))

root_tp_df = pd.DataFrame(root_tp_dict)
root_tp_df
```

```python
########
#### Create Master Stim Sheet
########


final_stims_dict = {'word': [],
                    'lemma': [],
                    'lemma_log_freq': [],
                    'RE': [],
                    'word_log_freq': [],
                    'root_a1a2_TP': [],
                    'root_a1_TP': [],
                    'a1_a2_TP': [],
                    'root_a2_TP': []}


def create_all_forms(all_stim_lemmas):
    words_array = []
    lemma_array = []
    endings_array = []
    
    for lem in all_stim_lemmas:
        if lem.endswith("ar"):
            endings = ["abas", "ábamos", "aban", "arías", "aríamos", "arían"]
        elif lem.endswith("er"):
            endings = ["ías", "íamos", "ían", "erías", "eríamos", "erían"]
        elif lem.endswith("ir"):
            endings = ["ías", "íamos", "ían", "irías", "iríamos", "irían"]
        
        words_array.extend([lem[:-2] + ending for ending in endings])
        lemma_array.extend(lem for x in range(6))
        endings_array.extend([ending for ending in endings])
        

    return(words_array, lemma_array, endings_array)

# Example usage:
# test = ["bailar", "cenar", "partir", "comer"]
all_stim_words, all_stim_lemmas, all_stim_endings = create_all_forms(all_stims_dict["lemma"])


counter = 0
for w in all_stim_words:
    cur_lemma = all_stim_lemmas[counter]
    # print(w)
    # print(cur_lemma)
    # print(all_stims_df[all_stims_df['lemma'] == cur_lemma]['lemma_freq'].iloc[0])
    # print(all_stims_df[all_stims_df['lemma'] == cur_lemma]['RE'].iloc[0])
    # print(0)
    # print(get_TP(cur_lemma, w))
    # print("____")
    final_stims_dict['word'].append(w)
    final_stims_dict['lemma'].append(cur_lemma)
    final_stims_dict['lemma_log_freq'].append(all_stims_df[all_stims_df['lemma'] == cur_lemma]['lemma_freq'].iloc[0])
    final_stims_dict['RE'].append(all_stims_df[all_stims_df['lemma'] == cur_lemma]['RE'].iloc[0])
    final_stims_dict['word_log_freq'].append(get_word_p(w, cur_lemma[-2:]))
    final_stims_dict['root_a1a2_TP'].append(get_TP(cur_lemma, w))

    # Root to {affix1, affix2} transition probabilities must be added based on tense 
    if w.endswith(("abas", "ías")) and not w.endswith(("arías", "erías", "irías")):
        final_stims_dict['root_a1_TP'].append(root_tp_df[root_tp_df['lemma'] == cur_lemma]['imperfect_stem_a1'].iloc[0])
        final_stims_dict['root_a2_TP'].append(root_tp_df[root_tp_df['lemma'] == cur_lemma]['imperfect_stem_a2_s'].iloc[0])

    elif w.endswith(("ábamos", "íamos")) and not w.endswith(("aríamos", "eríamos", "iríamos")):
        final_stims_dict['root_a1_TP'].append(root_tp_df[root_tp_df['lemma'] == cur_lemma]['imperfect_stem_a1'].iloc[0])
        final_stims_dict['root_a2_TP'].append(root_tp_df[root_tp_df['lemma'] == cur_lemma]['imperfect_stem_a2_mos'].iloc[0])
        
    elif w.endswith(("aban", "ían")) and not w.endswith(("arían", "erían", "irían")):
        final_stims_dict['root_a1_TP'].append(root_tp_df[root_tp_df['lemma'] == cur_lemma]['imperfect_stem_a1'].iloc[0])
        final_stims_dict['root_a2_TP'].append(root_tp_df[root_tp_df['lemma'] == cur_lemma]['imperfect_stem_a2_n'].iloc[0])

    
    elif w.endswith(("arías", "erías", "irías")):
        final_stims_dict['root_a1_TP'].append(root_tp_df[root_tp_df['lemma'] == cur_lemma]['conditional_stem_a1'].iloc[0])
        final_stims_dict['root_a2_TP'].append(root_tp_df[root_tp_df['lemma'] == cur_lemma]['conditional_stem_a2_ias'].iloc[0])
        
    elif w.endswith(("aríamos", "eríamos", "iríamos")):
        final_stims_dict['root_a1_TP'].append(root_tp_df[root_tp_df['lemma'] == cur_lemma]['conditional_stem_a1'].iloc[0])
        final_stims_dict['root_a2_TP'].append(root_tp_df[root_tp_df['lemma'] == cur_lemma]['conditional_stem_a2_iamos'].iloc[0])
        
    elif w.endswith(("arían", "erían", "irían")):
        final_stims_dict['root_a1_TP'].append(root_tp_df[root_tp_df['lemma'] == cur_lemma]['conditional_stem_a1'].iloc[0])
        final_stims_dict['root_a2_TP'].append(root_tp_df[root_tp_df['lemma'] == cur_lemma]['conditional_stem_a2_ian'].iloc[0])

    # Affix1 to Affix2 probabilities are based on the suffix
    final_stims_dict['a1_a2_TP'].append(a1_a2_TP_df[a1_a2_TP_df['suffix'] == all_stim_endings[counter]]['TP'].iloc[0])

    counter += 1

final_stims_dict
final_stims_df = pd.DataFrame(final_stims_dict)
final_stims_df
```
